Q3:

Which job roles are most common in the AI field?
● Find the top 10 most frequent job titles (e.g., AI Engineer, ML Scientist, LLM
Specialist). Create a bar chart for visualization.

```python
#df.mode() → most frequent value TO see the most frequent number
```

```python
## SUMMARIZING STATISTICS:

job_title_counts = df['job_title'].value_counts()  # counting how many unique values we have.

# Select the top 10 most frequent job titles:

most_frequent_jobs = job_title_counts.head(10)

# COMINING SEABORN AND MATPLOTLIB:

sns.set_style('ticks')  # makes it more readable and uniform.
plt.figure(figsize=(12, 6)) # plt.figure to create a figure object.
sns.barplot(x=most_frequent_jobs.index, y=most_frequent_jobs.values, palette='plasma')


plt.title('Top Most Common AI Jobs in 2025') # this is the head of the bar.
plt.xlabel('JOB TITTLES')  # bottom of the bar chart
plt.ylabel('JOBS AVAILABLE')  # side y axis of the chart
plt.xticks(rotation=20, ha='right') ## adjusting to improve uniform at x index
plt.show()
#seaborn and matplotlib being used to create a bar chart above^
```

9. Has AI hiring changed over time in 2025?
● Use Posted_Date to analyze trends are postings increasing or decreasing
month-to-month? Create a time-series chart showing job post freque

```python
df['posted_date'] = pd.to_datetime(df['posted_date'])
```

```python
df['year_month'] = df['posted_date'].dt.to_period('M')
display(df.head())
```

```python
monthly_counts = df.groupby('year_month').size()
display(monthly_counts)
```

```python
monthly_counts = monthly_counts.sort_index()
display(monthly_counts)
```

```python
plt.figure(figsize=(15, 7))
sns.lineplot(x=monthly_counts.index.astype(str), y=monthly_counts.values)
plt.title('Monthly AI Job Postings Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```
